## 2. 데이터셋 불러오기

`load_breast_cancer()`를 사용하면 Colab에서 바로 데이터셋을 불러올 수 있습니다.

## 3. 입력 데이터와 정답 확인하기

`data.data`에는 입력 feature들이 들어 있고, `data.target`에는 정답 label이 들어 있습니다.

## 4. Feature 이름과 Target 이름 확인하기

이 데이터셋의 feature는 종양의 반지름, 질감, 둘레, 면적 등 여러 측정값입니다.

````python
print("Feature names:")
print(data.feature_names)

print("\nTarget names:")
print(data.target_names)
````

## 5. 데이터프레임으로 확인하기

배열 형태보다 표 형태로 보면 데이터를 이해하기 쉽습니다.

````python
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

df.head()
````

## 6. Target 분포 확인하기

이진분류 문제에서는 각 클래스의 데이터 개수가 어느 정도 균형적인지 확인하는 것이 중요합니다.

````python
print(df["target"].value_counts())

plt.figure(figsize=(5, 4))
df["target"].value_counts().sort_index().plot(kind="bar")
plt.xticks([0, 1], data.target_names, rotation=0)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Target Distribution")
plt.grid(axis="y")
plt.show()
````

## 7. 기초 통계량 확인하기

feature들의 값 범위가 서로 많이 다릅니다.

예를 들어 면적 관련 feature는 값이 크고, smoothness 관련 feature는 값이 작습니다. 따라서 모델 학습 전에 scaling이 필요할 수 있습니다.

# GPT 답변을 검토할 때 체크할 것

## 1. 문제 유형이 이진분류인가?

이번 문제는 0 또는 1을 예측하는 binary classification입니다.

## 2. Loss function이 적절한가?

이진분류에서는 보통 다음 둘 중 하나를 사용합니다.

### 방법 A: BCEWithLogitsLoss 사용

이 방법은 마지막 layer에서 sigmoid를 직접 쓰지 않습니다.

```python
criterion = nn.BCEWithLogitsLoss()
```

평가할 때는 sigmoid를 적용합니다.

```python
probs = torch.sigmoid(outputs)
preds = (probs >= 0.5).float()
```

### 방법 B: BCELoss 사용

이 방법은 모델 마지막에 sigmoid를 넣어야 합니다.

```python
criterion = nn.BCELoss()
```

초보자 실습에서는 **BCEWithLogitsLoss 방식이 더 안정적**입니다.

## 3. 출력 크기가 1인가?

```python
nn.Linear(hidden_dim, 1)
```

## 4. y shape이 맞는가?

```python
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
```

## 5. CrossEntropyLoss를 쓴다면 출력 크기가 달라야 합니다

`CrossEntropyLoss`를 사용하려면 출력 크기를 2로 만들고, y는 long type의 class index여야 합니다.

```python
nn.Linear(hidden_dim, 2)
criterion = nn.CrossEntropyLoss()
```
