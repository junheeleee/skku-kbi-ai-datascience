# Data Preprocessing(sklearn, pandas)

```python
import pandas as pd

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing.DESCR)
```

```python
x_data = housing.data
y_data = housing.target
print(x_data.shape, y_data.shape)   # (506, 13) (506,)
```

```python
housing.data
```

```python
import pandas as pd
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df.head()
```

```python
df['target'] = y_data
df.head()
```

```python
df.describe()
```

```python
unit = 1000

plt.hist(df['target'])               # target 값의 분포를 히스토그램으로 그림
plt.legend(['MedInc'])              # 범례 설정 (MedInc로 이름표)
plt.ylabel('{}$'.format(unit))      # y축 라벨에 '1000$' 표시
plt.show()                          # 그래프 출력
```

## Data Preprocessing(Normalization, Standardization)

```python
def func_minmaxscale (number, col):
    mx = x_data[:,col].max()
    mn = x_data[:,col].min()
    # (number - min) / (max - min)  -> [0, 1]
    return (number-mn)/(mx-mn)

func_minmaxscale(x_data[0,2],2)
```

sklearn document 웹페이지 : https://scikit-learn.org/stable/modules/classes.html
https://scikit-learn.org/stable/modules/classes.html#module-sklearn.preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
x_scaler = MinMaxScaler()

# fit 메서드 : 데이터에 대해서 fit
x_scaler.fit(x_data)
# transform 메서드: 목표로 하는 데이터를 minmax-scaling

x_data_scaled = x_scaler.transform(x_data)

df_minmax = pd.DataFrame(x_data_scaled, columns=housing.feature_names)
df_minmax['y'] = housing.target
df_minmax.head()
```

```python
def func_std(data):
    m = data.mean()
    s = data.std()

    return (data-m)/s

print(x_data_scaled[:5, 0])
print(func_std(x_data_scaled[:5, 0]))
```
