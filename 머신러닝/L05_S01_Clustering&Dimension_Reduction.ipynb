## **1-1. KMeans Clustering**

> **데이터 로드**

```python
# Load dataset
from sklearn.datasets import load_iris
iris = load_iris()

data = iris["data"]
feature_name = iris["feature_names"]
```

> **클러스터링**

* Iris 데이터셋은 Setosa, Versicolor, Virginica 품종으로 구성

```python
from sklearn.cluster import KMeans
# Load K-Means model from sklearn
# n_clusters parameter를 통해 클러스터 개수 지정
kmeans = KMeans(n_clusters=3)

# Make prediction after training
predict = kmeans.fit_predict(data)
```

```python
import numpy as np

unique_predictions = np.unique(predict)
print("predict에 포함되는 값:", unique_predictions)
```

### **PCA**

- 기존 데이터의 분산을 최대화하는 새로운 축(주성분, Principal Component)을 찾아 데이터를 재배치하여 차원을 축소

```python
from sklearn.decomposition import PCA

# PCA 모델을 생성
pca = PCA(n_components=2) # n_components= 주성분 갯수
pca.fit(digits.data)      # PCA 적용, Returns the instance itself.
```

```python
# 처음 두 개의 주성분으로 숫자 데이터를 변환
digits_pca = pca.transform(digits.data) # PCA를 데이터에 적용
digits_pca
```

```python
# PCA로 차원 축소된 데이터 'digits_pca'의 형태 확인
print(f"PCA 이전 데이터의 형태: {digits.data.shape}")
print(f"PCA 데이터의 형태: {digits_pca.shape}")
```

```python
print(f"PCA 이전 데이터의 타입: {type(digits.data)}")
print(f"PCA 데이터의 타입: {type(digits_pca)}")
```

```python
# 색상 코드 정의
colors = ['#476A2A', '#7851B8', '#BD3430', '#4A2D4E', '#875525',
              '#A83683', '#4E655E', '#853541', '#3A3120', '#535D8E']


for i in range(len(digits.data)):
    # 숫자 텍스트를 이용해 산점도 그리기
    plt.text(digits_pca[i, 0], digits_pca[i, 1], str(digits.target[i]), # x, y, 그룹; str은 문자로 변환
             color=colors[digits.target[i]], # 산점도 색상
             fontdict={'weight':'bold', 'size':9}) # font 설정

plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max()) # 최소, 최대
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max()) # 최소, 최대
plt.xlabel('first principle component') # x 축 이름
plt.ylabel('second principle componet') # y 축 이름
plt.show()
```

### **t-SNE**
* t-SNE는 비선형 차원 축소 기법으로, 데이터 간 거리와 유사성을 고려하여 고차원 데이터를 2, 3차원 등으로 줄여 가시화하는데에 유용하게 사용
* PCA vs. T-SNE: https://amazelimi.tistory.com/13

```python
### TSNE모델에는 transform 메소드가 없고 fit_transform만 있음
from sklearn.manifold import TSNE

# t-SNE 모델 생성 및 학습
tsne = TSNE(n_components=2, random_state=1213)
digits_tsne = tsne.fit_transform(digits.data)
digits_tsne
```

```python
# t-SNE로 차원 축소된 데이터 'digits_tsne'의 형태 확인
print(f"t-SNE 이전 데이터의 형태: {digits.data.shape}")
print(f"t-SNE 데이터의 형태: {digits_tsne.shape}")
```

```python
print(f"t-SNE 이전 데이터의 타입: {type(digits.data)}")
print(f"t-SNE 데이터의 타입: {type(digits_tsne)}")
```

```python
# 시각화
for i in range(len(digits.data)): # 0부터  digits.data까지 정수
    plt.text(digits_tsne[i, 0], digits_tsne[i, 1], str(digits.target[i]), # x, y , 그룹
             color=colors[digits.target[i]], # 색상
             fontdict={'weight': 'bold', 'size':9}) # font

plt.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max()) # 최소, 최대
plt.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max()) # 최소, 최대
plt.xlabel('t-SNE Feature1') # x축 이름
plt.ylabel('t-SNE Feature2') # y축 이름
plt.show() # 그래프 출력
```

# **3. Outlier Detection**

## **3-1. K-Means Outlier Detection**
* 비지도 학습을 사용한 이상치 탐지
* 대표적인 비지도 학습 모델인 K-means 모델을 활용하여 중심에서 멀리 떨어진 데이터를 이상치로 탐지

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 랜던 2D 샘플 데이터 생성
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)   # K-Means 모델 학습

labels = kmeans.labels_  # 각 데이터의 군집 라벨
centroids = kmeans.cluster_centers_  # 군집 중심 좌표
distances = np.linalg.norm(X - centroids[labels], axis=1) # 데이터와 군집 중심 간 거리
```

```python
threshold = np.mean(distances) + 2.5 * np.std(distances)
outliers = X[distances > threshold]
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200,
            c='red', label="Centroids")
plt.scatter(outliers[:,0], outliers[:,1], marker='o', s=100,
            edgecolors='r', facecolors='none', label="Outliers")
plt.legend()
plt.title("K-Means Outlier Detection")
plt.show()
```
