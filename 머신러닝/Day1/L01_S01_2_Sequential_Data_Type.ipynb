# **Python: Sequential Data Type**

: 여러 개의 데이터를 하나의 변수에 가지고 있는 데이터 타입

**< Sequence Type >** (동일한 operation 적용 가능 (indexing / slicing))
- List(리스트)
- Tuple(튜플)
- String(문자열)

**< Set Type >**
- Set(집합)

**< Mapping Type - Associative array >**; Hash table
- Dictionary(딕셔너리)



## **1. List(리스트)**

* 리스트에는 다양한 타입의 원소(문자열, 숫자, 리스트 등등)를 모두 포함할 수 있음
* **순서가 있는(ordered)** **변경 가능한(mutable)** 데이터들의 집합


> **리스트 만드는 방법**

```python
# [] 사용
list0 = [1,'2',[3]]
print(list0)
```

```python
# 빈 리스트
list1 = []
print(list1)

list2 = list()
print(list2)
```

```python
# 2차원 리스트(2x2 행렬처럼 표현 가능)
list3 = [[1,2],[3,4]]
print(list3) # 실제로는 2개의 리스트를 원소로 갖는 리스트
```

> **리스트로 입력 받기**

```python
# 한 개씩 입력받기(엔터로 구분됨)
N = 3
list1 = []
for i in range(N) :
  list1.append(input())
print(list1)
```

```python
list2 = [int(input()) for i in range(3)] ## *[] 안에 for문 적용*
print(list2)
```

```python
# 여러개 입력받기(띄어쓰기로 구분됨)
list3 = list(input().split()) ## 문자형으로 받기
print(list3)
```

```python
list4 = list(map(int,input().split())) ## 정수형으로 받기
print(list4)
```

> ***Indexing (어떤 원소에 접근)**

```python
L = [1,2,3,4,5]
print(L)

# L의 첫번째 원소
print(L[0])

# L의 마지막 원소 ## 리스트 원소의 길이; len()
print(L[-1])
print(L[len(L)-1] )
```

```python
# indexing을 이용한 연산
print(L[1]+L[4])
```

```python
# 리스트 안의 리스트 인덱스
L1 = [1,[2,3],4]
print(L1[1][0]) # [2,3][0]
```

> ***Slicing (~부터 ~까지)**

```python
# slicing; list[start:end:step]
print(L[:3]) #=> L[0:3]
print(L[2:4])
print(L[1:]) #=> L[1:5], L[1:len(L)]
```

```python
## n칸씩 slicing; [::n]
print(L[::2])
```

> **리스트 연산**

```python
# 리스트 더하기
List1 = [1,2]
List2 = ['a','b','c']
print(List1+List2)
```

```python
# 리스트 곱하기
print(List1 * 3)
```

```python
# 리스트 수정하기 # 되돌릴 순 없음!
List1[0] = 20
print(List1)
```

```python
List2[1:] = ['v','i']
print(List2)
```

> **리스트 관련 함수**

```python
# 원소 추가하기; append()
lst1 = []
lst1.append(5)
lst1.append(1)
lst1.append(2)
print(lst1)
```

```python
# 해당 위치에 원소 삽입; insert(위치, 값)
lst1.insert(1,4)
print(lst1)
```

```python
# 원소 정렬하기; sort() - 알파벳, 한글 정렬 가능! (list에서만 사용 가능)
lst1.sort() # 오름차순
print(lst1)
lst1.sort(reverse=True) # 내림차순
print(lst1)
```

```python
# 해당 원소의 개수; count()
print(lst1.count(4))
```

```python
# 리스트 뒤집기-정렬x; reverse()
lst2 = [1,8,3,5,0]
lst2.reverse() # OR lst2[::-1]
print(lst2)
```

```python
lst2 = [1,8,3,5,0]

# 마지막 원소 제거; pop()
lst2.pop()
print(lst2)

# 해당 위치 원소 제거; pop(위치)
lst2.pop(2)
print(lst2)
```

```python
# 해당 원소 제거; remove() - 중복될 경우 첫번째 원소만 제거
lst2.remove(8)
print(lst2)
```

```python
# 해당 위치 원소 제거; del list[]
del lst2[0]
print(lst2)
```

```python
# 리스트 요소의 위치 반환; index(value)
print(lst1.index(2))
```

## **2. Tuple(튜플)**

* **순서가 있는(ordered) 변경 불가능한(immutable)** 데이터들의 집합
* 다양한 타입의 데이터를 저장할 수 있음
* 한 번 생성되면 원소 추가, 삭제, 수정 불가능 (데이터 무결성 유지가 필요할 떄 사용)

> **튜플 만드는 방법**

```python
# () 사용
t0 = (1,'2',[3],(4,5))
print(t0)

# 빈 튜플
t1 = ()
print(t1)

t2 = tuple()
print(t2)
```

> ***Indexing (어떤 원소에 접근)**

```python
t = (1,3,5,7,9)
print(t)

# t의 첫번째 원소
print(t[0])

# t의 마지막 원소 ## 리스트 원소의 길이; len()
print(t[-1])
print(t[len(t)-1] )

# indexing을 이용한 연산
print(t[1]+t[4])
```

> ***Slicing (~부터 ~까지)**

```python
# slicing; tup[start:finish:step]
print(t[:3]) #=> L[0:3]
print(t[2:4])
print(t[1:]) #=> t[1:5], t[1:len(t)]

## n칸씩 slicing; [::n]
print(t[::2])
```

> **튜플 연산**

```python
# 튜플 더하기
tup1 = (1,2)
tup2 = ('a','b','c')
print(tup1 + tup2)

# 튜플 곱하기
print(tup1 * 3)
```

## **3. Set(집합)**

* **순서가 없고(unordered) 중복을 허용하지 않는(unique)** 데이터들의 집합
* 데이터에서 중복을 제거하거나 수학적 집합 연산을 수행할 때 사용
  * 공집합을 생성할 시, set()로 생성 ({} : 빈 dict 생성)
  * 집합의 연산자 : 교집합, 합집합, 차집합


> **집합 만드는 방법**

```python
# {} 사용 OR set([])
s0 = {1,(4,5),'2'} ## 리스트는 원소로 포함될 수 없음!
# s0 = set([1,'2',(4,5)])
print(s0)


# 빈 집합(공집합)
s1 = set()
print(s1)
```

```python
s = {}
type(s)
```

```python
# list or tuple -> set (*원소의 uniqueness 활용)
lt = {1,2,5,2,3,43,4,6,6,3,2,4}
print(set(lt)) # 중복 제거
```

> **집합 연산**

```python
s1 = {1,2,3,4,5}
s2 = {1,3,5,6,9}

# 교집합
print(s1 & s2)
print(s1.intersection(s2))
```

```python
# 합집합
print(s1 | s2)
print(s1.union(s2))
```

```python
# 차집합
print(s1 - s2)
print(s1.difference(s2))
print(s2 - s1)
```

> **집합 관련 함수**

```python
# 원소 하나 추가; add()
S = set()
S.add(8)
print(S)
```

```python
# 여러 원소 추가; update() - set 자체 변경
S1 = {4,8,48}
S1.update({3,6}) # S1.union({3,6}) or S1 | {3,6} - 새로운 set 생성
print(S1)
```

```python
# 원소 하나 제거; remove()
S1.remove(48)
print(S1)
# 모든 원소 제거; clear()
```

```python
# 최소/최대; min()/max()
S2 = {4, 6, 3, 10, 8}
print(max(S2))
print(min(S2))
```

## **4. Dictionary(사전)**

* 키(key)와 값(value)의 쌍으로 데이터를 저장 (Table Concept)
* index가 아닌 key를 통해 value에 접근
* key가 될 수 있는 데이터 타입은 변경 불가능(immutable) 해야 함
* 순서가 중요하지 않음

> **사전 만드는 방법**

```python
# {:} 사용
d0 = {'a':1, 'b':2}
print(d0)
```

```python
# 빈 사전
d1 = {}
print(d1, type(d1))

d2 = dict()
print(d2)
```

```python
### key 중복 허용X(뒷쪽에 있는 value 적용)
d3 = {'a':1, 'b':2, 'a':3}
print(d3)
```

> ***dict indexing**

```python
print(d0['a'])
```

```python
# d0에 key 'c'가 없는 경우
d0['c'] = 3 # -> 추가
print(d0)
```

```python
# d0에 key 'c'가 있는 경우
d0['c'] = 4 # -> 수정
print(d0)
```

> **사전 관련 함수**

```python
D = {'name':'yoon','birth':'0503','phone':'95107530','age':'23'}
```

```python
# 모든 key값들 보기; keys()
print(D.keys())

# 모든 value들 보기; values()
print(D.values())

# key, value 쌍으로 보기; items()
print(D.items())
```

```python
# 해당 value 가져오기; get()
print(D.get('name')) # D['name']
print(D.get('address', 'None')) ## key값이 없다면 ',' 뒤의 내용(디폴트값) 출력
```

```python
# 원소 삭제; del(D[key값]) or del D[]
del(D['birth'])
print(D)

# 원소 삭제 후 value 반환; pop(key값)
print(D.pop('age'))
```

```python
# 해당 key값이 존재하는지 확인; in ~ 모든 연속형 데이터 타입에서 사용 가능
print('phone' in D)
print('school' in D)
print('name' in D.keys())

# 해당 value가 존재하는지 확인; in dic.value()
print('yoon' in D.values())
```
