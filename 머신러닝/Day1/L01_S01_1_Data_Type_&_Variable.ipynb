### **1-1. Operator(연산자)**

: 데이터를 가지고 특정 작업을 수행할 때 사용하는 기호

#### **기본연산자**
* 산술
* 대입(=)
* 복합 대입

> **산술 연산자**: 숫자형 데이터가 제공하는 연산자

```
### 산술 연산자 우선순위 ###
1. ()   2. **  3. * / // %  4. + -
```



````python
a = 10
b = 5

print(a + b)    #덧셈

print(a - b)    #뺄셈

print(a * b)    #곱셈

print(a / b)    #실수 나누기

print(a // b)   #정수 나누기(몫)

print(a % b)    #나머지

print(a ** b)   #거듭제곱
````

> **대입 연산자**: 변수에 특정 값이나 계싼 결과를 저장하는 연산자

````python
a = 3
print(a)
````

> **복합 대입 연산자**: 산술 연산자와 대입 연산자를 결합해 계산 결과를 기존 변수에 바로 저장하는 연산자

````python
a += 3      #덧셈 a = a+3
print(a)
````

````python
a -= 3      #뺄셈 a = a-3
print(a)
````

````python
a *= b      #곱셈 a = a*b
print(a)
````

````python
a /= b      #실수 나누기
print(a)
````

````python
a //= b     #정수 나누기(몫)
print(a)
````

````python
a %= b      #나누기
print(a)
````

````python
a **= b     #거듭제곱
print(a)
````

#### **관계(비교)연산자**

: 두 값을 비교하여 True 또는 False 결과를 반환하는 연산자

````python
print(1 == 2)   # =는 '대입'을 의미하며, ==는 '같음'을 의미함

print(1 != 10)  # 같지 않음

print(1 > 2)    # 큼

print(1 >= 1)   # 크거나 같음

print(1 < 2)    # 작음

print(1 <= 0)   # 작거나 같음
````

> 관계 연산자의 경우, 문자열 비교도 가능!

* 문자열에 해당하는 유니코드를 비교

````python
print('cat'<'cog') # 알파벳 순

print('Dog'<'dog') # 대문자 < 소문자

print('Dog'<'cat')

print('알고리즘'>'파이썬') # 가나다 순

print('알고리즘'>'algorithm') # 알파벳 < 한글
````

````python
# 단, 숫자와 문자열 비교 X

print('알고리즘' < 10)
````

#### **논리연산자**
: 여러 조건식의 True/False를 통해 논리적인 결과를 판단하는 연산자

````python
A = 1
B = 10

# and : 둘 다 참이면 참
print(A > 0 and B > 0)
print(A > 0 and B > 100)
````

````python
# or : 둘 중 하나만 참이여도 참
print(A > 0 or B > 100)
````

````python
# not : 논리 상태 반전
print(not A > 2) # A > 2 => False
````

## **2. Variable(변수)**

* 변수는 데이터를 저장할 수 있는 이름이 붙은 공간
* 변수에 값을 할당하여 나중에 이 값을 다시 사용하거나 변경 가능
* **Assignment operator**: 변수에 데이터를 저장할 때 사용하는 `=`
* **lvalue**: =를 기준으로 왼쪽 (실제 메모리 주소)
* **rvalue**: =를 기준으로 오른쪽 (실제 데이터)

</br>

**<변수명 규칙 (권장)>**
- 영문자를 사용함
- 소문자로 시작함(클래스의 이름을 대문자로 시작하므로)
- 데이터의 의미를 쉽게 파악할 수 있도록 함
- 두 개 이상의 단어 => 낙타표기법(ex newItem)
- 언더바(\_)를 제외한 특수문자 사용x  
  언더바(\_)로 시작하면 특별한 의미를 가지게 되므로 주의!
- 공백문자( ) 사용x
- 첫 글자에 숫자가 올 수 없음

````python
a = 10
print(id(a)) # a라는 변수가 실제 메모리에서 저장되어있는 위치(메모리 주소)
````

````python
# 입력 함수; input() > 문자열 형태로 저장
ip = input('input값 : ')
print(ip, type(ip))
````

````python
ip1 = int(input('정수 input값 : ')) # 자료형 변환
print(ip1, type(ip1))
````

> **문자열 Formatting**

````python
# ',' 이용한 출력 >> 변수 앞뒤로 이용해야함, 띄어쓰기 포함
a = '사과'
b = '4'
print(a,'는',b,'개 있다.')
````

````python
# print format
## %s(str), %d(int), %f(float)
print('%s는 %s개 있다.' %(a,b))
````

````python
# str.format 함수
print('{}는 {}개 있다.'.format(a,b))
````

````python
# f-string
a = '사과'
b = 4
print(f'{a}는 {b}개 있다.')
````

> **문자열 메소드(관련 함수)**

````python
# 문자열의 길이를 출력하는 함수; len()
s = 'Hello World'
print(len(s))
````

````python
# 대소문자 바꾸기; upper(), lower()
print(s.upper()) # 대문자로 바꿈
print(s.lower()) # 소문자로 바꿈
````

````python
ss = s.lower()
ss
````

````python
## (문자 전체 or 글자 하나)소문자인지 대문자인지 boolean형태로 구분; isupper(), islower()
print(s.isupper())
print(s[0].islower())
````

````python
# 문자 공백 지우기; strip() # 단, 문자 사이의 공백은 지워지지 않음
s_st = '     h  i   '
print(s_st.strip())
````

````python
# 문자열 삽입; join()
print('+'.join('abcd'))

list1=['010','1234','1234'] # 리스트의 원소들 사이에도 삽입 가능
print('-'.join(list1))
````

````python
# 문자열 나누기; sprit()
print(s.split())
print(s.split('l')) # 'l'(token)을 기준으로 나눔

### 문자가 중복될 땐 이를 ''(공백표시)로 표현
````

````python
# 문자열 바꾸기; replace()
print(s.replace('World','Python'))
print(s.replace(' ','')) # 빈칸 없애기
````

````python
# 문자열 시작 or 끝 확인; startswith(), endswith()
print(s.startswith('Hell'))
print(s.endswith('b'))
````

````python
# 문자열에 속한 문자'x' 개수 반환; count(x)
print(s.count('l'))
````

````python
# 문자열에서 문자'x'가 처음 나타나는 인덱스 반환; find(x)
print(s.find('l'))
````
