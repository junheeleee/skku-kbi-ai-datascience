## **1. Conditional Statement**

* 특정 *조건*이 참(True)일 때만 코드를 실행하거나, 조건에 따라 다른 코드를 실행
* '만약 ~라면, ~하라'와 같은 형태로 프로그램의 의사결정을 표현

### **1-1. If**

* 가장 기본적인 조건문
* `if` 뒤에 오는 조건이 참(True)일 경우에만 `if` 블록 안의 코드가 실행
* 조건이 거짓(False)이면 해당 블록은 완전히 무시

```python
a = 10
b = 50

# if
if a == 10 :
  print('a')

if b == 30 :
  print('b') # 조건이 만족하지 않아 출력이 안됨
```

### **1-2. `if-else`**

* `if` 조건이 참(True)일 때와 거짓(False)일 때 각각 다른 코드를 실행하고 싶을 때 사용
* `if` 조건이 참이면 `if` 블록의 코드, 거짓이면 `else` 블록의 코드 실행

```python
# if - else
if a != 10 :
    print('False')
else :
  print('True')
```

> **삼항 연산자 (Conditional Expression)**

* 간단한 `if-else` 문은 한 줄로 축약하여 표현
* `참일_때_값 if 조건 else 거짓일_때_값` 형식으로 사용

```python
## 한 문장으로 조건문 사용 가능
i = 5
'조건 만족' if i > 10 else '조건 불만족'
```

> **예제**

```python
'''
어떤 연도를 N이라고 할 때, N이 윤년인지 아닌지 알아보자
[윤년 : 4의 배수이면서, 100의 배수가 아니거나 400의 배수]
'''

n = int(input('N : '))

if n%4 == 0 and n%100 != 0 or n%400 == 0:
    print('윤년')
else :
    print('윤년 아님')
```

### **1-3. `if-elif-else`**

* 여러 가지 조건을 순차적으로 검사하고 싶을 때 사용
* `if` 조건이 거짓일 경우 다음 `elif` 조건을 검사하고, 모든 `elif` 조건마저 거짓일 경우 마지막 `else` 블록의 코드가 실행

```python
'''
자판기에 입력된 돈이 300원이면 커피를 주고,
300원보다 적은 경우엔 돈을 그냥 돌려주고,
300원보다 많은 경우에는 커피와 함께 거스름돈 주기
'''
# if - elif - else
money = int(input('투입한 돈 : '))
if money < 300 :
  print('%d원을 돌려줍니다.' %(money))
elif money == 300 :
  print('coffee')
else :
  print('coffee')
  print('%d원을 돌려줍니다.' %(money-300))
```

> **중첩 `if` (Nested `if`)**

* 조건문 안에 또 다른 조건문을 사용
* 복잡한 상황에서 여러 단계의 조건을 처리해야 할 때 유용하게 사용
* `if-elif-else`와 동일한 결과

```python
# 중첩 if (nested if)
money = int(input('투입한 돈 : '))
if money == 300 :
  print('coffee')
else :
  if money < 300 :
    print('%d원을 돌려줍니다.' %(money))
  else :
    print('coffee')
    print('%d원을 돌려줍니다.' %(money-300))
```

## **2. Iteration Statement(반복문)**

* 특정 코드 블록을 여러 번 반복해서 시랳ㅇ할 떄 사용
* 파이썬에서는 `while`과 `for` 반복문 사용

### **2-1. While Statement (while문)**
* 조건이 만족할 때 까지 반복
* **초기값, 조건식, 증감값** 설정
* 조건이 항상 참이면 무한 루프(Infinite Loop)에 빠질 수 있으니 주의

```python
'''
커피가 다 떨어질 때까지 자판기 프로그램을 반복하면서 실행한 뒤,
돈을 입력받아서 입력된 돈이 300원이면 커피를 주고,
300원보다 적은 경우엔 돈을 그냥 돌려주고,
300원보다 많은 경우에는 커피와 함께 거스름돈 주기
'''
t_cf = 3

while t_cf > 0 :
  money = int(input('투입한 돈 : '))
  if money < 300 :
    print('%d원을 돌려줍니다.' %(money))
  elif money == 300 :
    print('coffee')
    t_cf -= 1
  else :
    print('coffee')
    print('%d원을 돌려줍니다.' %(money-300))
    t_cf -= 1

print('커피가 모두 소진되었습니다.')
```

### **2-2 For Statement(for문)**

* 지정 횟수동안 반복 수행
* 리스트, 튜플, 문자열과 같은 반복 가능한(iterable) 객체의 모든 요소를 순회하여 코드를 실행할 때 사용
* 반복 횟수나 순회할 대상이 명확할 때 더 직관적이고 편리

> **for문과 list**

* 리스트의 각 요소를 순서대로 변수에 할당하면서 반복

```python
# for 변수 in 리스트
coffMenu = ['아메리카노','카페 라떼','카페 모카','카라멜라떼','바닐라라떼','콜드브루']

for i in coffMenu :
  print(i)
```

> **for 문과 `range()` 함수**

* `range()`: 일련의 숫자를 생성
* `for 변수 in range(start, end+1, step):`: `start`부터 `end` 숫자까지 `step`만큼 증가하는 숫자를 `변수`에 할당하면서 반복
* `step`는 생략 가능하며, 생략 시 기본값은 1
* `start`도 생략 가능하며, 생략 시 기본값은 0

```python
# for 변수 in range(start,end+1,step)
for j in range(3,10,3) :
  print(j)
```

> **예제**

```python
'''
6개의 커피 음료와 음료의 가격이 담긴 리스트가 있다.
가지고 있는 돈이 5000원일 때 먹을 수 있는 모든 음료를 찾아보기.
'''
coffMenu = ['아메리카노','카페 라떼','카페 모카','카라멜라떼','바닐라라떼','콜드브루']
price = [4400,4800,5000,5300,5300,4600]

for k in range(len(coffMenu)) :
  if price[k] <= 5000 :
    print(coffMenu[k])
```

> **`for` 문을 이용한 리스트 값 변경**

* `range(len(price))``를 통해 인덱스에 접근한 뒤 값을 직접 수정할 수 있음

```python
## 음료 500원 할인
for n in range(len(price)) :
  price[n] -= 500
print(price)
```

> **`print()` 함수의 `end` 파라미터**

* 파이썬의 `print()` 함수는 기본적으로 출력 후 줄바꿈
* `end` 파라미터를 사용하면 줄바꿈 대신 원하는 문자열로 끝낼 수 있음

```python
# 줄을 바꾸지 않고 한 줄에 출력하기; end=''
for i in range(1,6) :
  print(i,end=' ')
```

### **2-3. 중첩 `for` 문 (Nested `for` Loop)**

* `for` 문 안에 또 다른 `for` 문을 사용

```python
# 중첩 for문 (nested loop)
for dan in range(2,10) :
  for i in range(1,10) :
    print('%d x %d = %d' %(dan, i, dan*i))
  print('-'*10)
```

### **2-4. 반복문 제어 (Control Statements for Loops)**

* `break` : 반복문을 수행하다가 더 이상 반복이 필요없는 경우 사용
* `continue` : 반복문을 수행하다가 특정 조건에만 건너뛰는 경우 사용
* `pass` : 해당 조건문 건너뛰기(수행할 문장이 없음)

```python
# break 예제
for i in range(5):
    if i == 3:
        break  # i가 3이면 반복문 종료
    print("break 예제:", i)
```

```python
# continue 예제
for i in range(5):
    if i == 2:
        continue  # i가 2일 때는 아래 문장 건너뛰기
    print("continue 예제:", i)
```

```python
# pass 예제
for i in range(5):
    if i == 1:
        pass  # 아무 것도 하지 않음, 자리 채우기용
    print("pass 예제:", i)
```

> **무한 루프 (`while True`)와 `break`, `continue` 활용**

* `while True`는 조건이 항상 참이므로 무한히 반복되는 루프
* `break` 문을 사용하여 특정 조건에서 강제로 종료

```python
# 무한 루프(Infinite Loop); while True
t_cf = 3
extMoney = 0

while True :
  money = int(input('투입한 돈 : '))
  money = money + extMoney
  if money == -1 :
    print('운영을 종료합니다.')
    break
  if money < 300 :
    print('돈이 부족합니다. 추가로 금액을 투입해주세요.')
    extMoney = money
    continue
  elif money == 300 :
    print('coffee')
    t_cf -= 1
  else :
    print('coffee')
    print('%d원을 돌려줍니다.' %(money-300))
    t_cf -= 1
  extMoney = 0
  if t_cf == 0 :
    print('커피가 모두 소진되었습니다.')
    break
```
