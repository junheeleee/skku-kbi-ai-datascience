# **Python: Function**

* 특정 작업을 수행하는 코드 블록
* 정해진 순서대로 특정 작업을 수행하도록 만들어진 작은 프로그램 단위 (input -> output)
* 파이썬에서는 `def` 키워드를 사용

</br>

**함수(Function)의 장점**
* **재사용성**: 같은 코드를 여러 번 반복하여 작성할 필요 없이, 한 번 정의해두고 필요할 때마다 호출하여 사용할 수 있음
* **모듈성**: 복잡한 프로그램을 여러 개의 작은 함수로 나누어, 코드의 가독성을 높이고 유지보수를 쉽게
* **오류 방지**: 기능별로 코드를 분리하여 오류 발생 시 문제의 원인을 빠르게 찾고 해결하기 용이


## **1. Function Definition(함수 구현)**

### **1-1. 매개변수 (Parameter), 인수(Argument)**
: 함수가 외부로부터 전달받는 값들을 저장하는 변수 (input 값)

```python
def func(para1, para2) :
  a = para1 + para2
  return a

func(1,2)
```

```python
def summ(a, b) :
  return a + b

def sub(a, b) :
  return a - b

a = 5
b = 10

print('sum :',summ(a,b))
print('sub :',sub(a,b))
```

> **parameter O, return O**

```python
def sum(a, b) :
  return a + b

a = 1
b = 10

sum(a, b)
```

> **parameter X, return O**

```python
def say_hello():
    return "Hello, World!"

print(say_hello())

message = say_hello()
print(message)
```

> **parameter O, return X**

```python
def my_name(name) :
  print('My name is',name)

my_name("Python")
```

> **parameter X, return X**

```python
def say_hi() :
  print('Hi!')

say_hi()
```

### **1-2. 매개변수 더 알아보기**

> **가변인자(Arbitrary Arguments: `*args`)**:
* 함수가 임의의 개수의 인수를 튜플 형태로 받아들일 수 있음
* 함수를 호출할 때 몇 개의 인수를 전달할 지 미리 알 수 없을 때 유용

```python
# 입력 parameter의 개수를 모르는 경우 -> 입력된 parameter은 tuple로 변환됨
def add_many(*args) :

  print(args)

  total = 0
  for num in args :
    total += num
  return total

add_many(1, 2, 3, 4, 5)
```

```python
add_many(1,3,5)
```

```python
# Keyword parameter; parameter의 수가 매우 많아 입력 parameter를 정하고 싶을 때
def df_add(a, c, b=10) :  # b : default parameter
  return a + b + c

print(df_add(1, 2, 3))
print(df_add(5, 15))
```

> **Parameter 범위**
* **Local variable (지역 변수)** : 함수 내에서만 유효
* **Global varialbe (전역 변수)** : 함수 밖에서 유효

```python
x = 10
x = 20
x
```

```python
x = 10  # 전역변수

def my_function():
    x = 5  # 지역변수
    print("지역변수 x:", x)

my_function()
print("전역변수 x:", x)
```

```python
def change_name(name):
    print(f"변경된 이름: {name}")

result = change_name("Alice")
print(result)  # None 출력
```

## **2. 파이썬 내에 정의된 함수**

* 파이썬은 기본적으로 제공하는 내장 함수와 다양한 기능을 가진 모듈들을 통해 프로그래밍을 더욱 편리하게 할 수 있음

```python
# 수학 연산
import math   # math; "library"

print(abs(-3)) # 절댓값
print(math.ceil(3.5)) # 반올림
print(math.floor(3.2)) # 반내림

print(math.sin(3.14)) # sin
print(math.cos(0)) # cos
```
